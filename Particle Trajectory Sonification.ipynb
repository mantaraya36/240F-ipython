```python
n = 30000
dt = 1e-1

r = zeros((n,2))
v = zeros((n,2))
a = zeros((n,2))

r[0] = [256, 256]
v[0] = [0, -2.0]

damp = 0.9999

q = 5500
m = 0.5

for i in range(n - 1):
    E = array([Ex[int(r[i][0]), int(r[i][1])], Ey[int(r[i][0]), int(r[i][1])]])

    a[i+1] = q*E/m 
    v[i+1] = damp * (v[i] + a[i+1]*dt)
    r[i+1] = r[i] + v[i+1]*dt
    
```

```python
figure(figsize=(12,10))
imshow(gspace.T)
#quiver(Ex[::],Ey[::])
plot(*r.T, lw=3, c= 'w')
xlim((0, 512))
ylim((0, 512))
```
