##A short digression into python performance

Thanks Kee!

```python
n = num_frames/len(data)
```

```python
%%timeit
phs_mod = array(())

for f in freq:
    phs_mod = r_[phs_mod, ones(n) * f]
```

```python
%%timeit
phs0 = []

for f in freq:
    temp = []
    for i in range(n):
        temp.append(f)
    phs0 += temp
```

```python
%%timeit
phs1 = []
for f in freq:
    temp = [f] * n
    phs1 += temp
```

```python
%%timeit

phs2 = [0] * n * len(freq)

for i, f in enumerate(freq):
    offset = i * n
    for j in range(n):
        phs2[offset + j] = f
```

```python
%%timeit
phs3 = [0] * (n * len(freq))

for i in range(len(phs3)):
    phs3[i] = freq[i//n]
```

```python
%%timeit
phs4 = [freq[i // n] for i in range(n * len(freq))]
```

```python
%%timeit
phs_mod = zeros(len(freq) * n)

index = 0
for f in freq:
    phs_mod[index:index + n] = f
    index += n
```

```python
phs_mod1 = array(())

for f in freq:
    phs_mod1 = r_[phs_mod1, ones(n) * f]

phs_mod2 = zeros(len(freq) * n)

index = 0
for f in freq:
    phs_mod2[index:index + n] = f
    index += n
    
allclose(phs_mod1,phs_mod2)
```
